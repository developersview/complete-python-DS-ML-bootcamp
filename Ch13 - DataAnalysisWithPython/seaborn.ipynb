## Data Visualization With Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#basic plotting with Seaborn
tips = sns.load_dataset('tips')
tips
```

```python
## create a scatter plot
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot of Total Bill vs Tips")
plt.show()
```

```python
## Line Plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total bill by size")
plt.show()
```

```python
##barplot
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("Bar Plot Total Bill vs day")
plt.show()
```

```python
#box plot
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
#Violin plot
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
#histogram
sns.histplot(tips['total_bill'], bins=10, kde=True)
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'], fill=True)
```

```python
#Pairplot
sns.pairplot(tips)
```

```python
tips
```

```python
#heatmap
corr = tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr, annot=True)
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```

```python

df=pd.read_csv('sales_data.csv')
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Product Category', y='Total Revenue', data=df)
plt.title("Bar Plot of Total Revenue as per Product Category")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Total Revenue', data=df)
plt.title("Bar Plot of Total Revenue as per Product Category")
plt.show()
```

```python
sns.histplot(df['Total Revenue'], bins=30, kde=True)
```

```python
#heatmap
corr = df[['Units Sold', 'Unit Price', 'Total Revenue', 'Transaction ID']].corr()
corr
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
```
